````python
f = 1
Ts = 0.01
A = 1
t = np.arange(0, 1+Ts, Ts)
x1 = A * np.sin(2 * np.pi * f * t)
x2 = A * np.cos(2 * np.pi * f * t)
plt.plot(t, x1)
plt.plot(t, x2)
plt.xlim(0, 1)
````

````python
f = 1
Ts = 0.01
A = 1
t = np.arange(0, 1+Ts, Ts)
x1 = A * np.sin(2 * np.pi * f * t)
x2 = A * np.cos(2 * np.pi * f * t)
plt.plot(t, x1, label="sine")
plt.plot(t, x2, label="cosine")
plt.xlim(0, 1)
plt.legend()
plt.xlabel("time($t$)", loc='right')
plt.ylabel("mag.")
plt.title("Sine/Cosine graph")
plt.savefig("rrrrrgrp.png")
````
