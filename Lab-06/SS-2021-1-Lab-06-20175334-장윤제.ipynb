#### 문제 2.

문제 1의 악보를 연주하는 프로그램을 완성하라.

```python
import numpy as np
from IPython.display import Audio
import copy
```

```python
def single_tone(duration, A, f, phi, fs=44100):#A, f, fs, phi, duration  #f = 주파수 fs = 샘플링주파수
    t = np.arange(0, duration, 1/fs)
    return t, A * np.sin(2 * np.pi * f * t + phi)
```

```python
def get_exp_decay_window(duration, a=10,fadeout=True, fs=44100):
    def fadeout_weight():
        nonlocal t,x
        x = x * (1-t/duration)
    Ts = 1/fs
    t = np.arange(0,duration,Ts)
    x = t*np.exp(-a*t)
    if fadeout:
        fadeout_weight()
    return t,x/max(x)
```

```python
fs = 44100
```

```python
def music_wave(music_score, key_no, tempo=60, fs=44100):
    wav = np.array([])
    for note in music_score:
        key = note[0]
        no = key_no[key]
        length = note[1] * (60/tempo)
        f = 440.0 * (2 ** ((no-49)/12))
        t, tone = single_tone(length, 0.8, f, 0)
        t, window = get_exp_decay_window(length,1)
        wav = np.append(wav, tone*window)
    return wav


upup_wave = music_wave(upup, key_no, tempo = 60, fs=44100)
updown_wave = music_wave(updown, key_no, tempo = 60, fs=44100)
downup_wave = music_wave(downup, key_no, tempo = 60, fs=44100)
downmiddle_wave = music_wave(downmiddle, key_no, tempo = 60, fs=44100)
downdown_wave = music_wave(downdown, key_no, tempo = 60, fs=44100)
downdownst_wave = music_wave(downdownst, key_no, tempo = 60, fs=44100)
```

```python
Audio((upup_wave+updown_wave)+(downup_wave+downmiddle_wave+downdown_wave+downdownst_wave), rate=fs, autoplay=True)
```
