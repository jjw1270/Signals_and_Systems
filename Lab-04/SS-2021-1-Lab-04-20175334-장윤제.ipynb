#### P-3.8 다음 방정식의 해를 구하고, 복소 평면에 모든 해를 표시하라.


(3) $z^8 = 1$

의 해는 모두 8개로 $z_{n} = e^{j\frac{2\pi}{8}n}$, n = 0,1,2,3,4,5,6,7 이다.
  
  $z = 1 , -1 , -j , j , e^{-\frac{i\pi}{4}} , e^{\frac{i\pi}{4}} , e^{-\frac{3i\pi}{4}} , e^{\frac{3i\pi}{4}}$


```python
# P-3.8 Python을 이용한 풀이
import matplotlib.pyplot as plt
import math
from sympy import *
import numpy as np

z = Symbol('z')


z = solve(z**8-1,z)
disp.display(z)

for n in range(len(z)):
    
    z[n] = complex(z[n])




real = [(n).real for n in z]
imag = [(n).imag for n in z]


a = plt.axes(xlim=(-2,2),ylim=(-2,2))
a.set_aspect('equal')


plt.scatter(real,imag)


plt.ylabel("Imaginary")
plt.xlabel("Real")
plt.axvline(x=0, color = 'black')
plt.axhline(y=0, color = 'black')
plt.show()
```
