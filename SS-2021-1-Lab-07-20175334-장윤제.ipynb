### 이산 푸리에 급수(DFS: discrete Fourier series)

- 주기가 $N$인 이산 주기 신호의 한 주기 동안의 샘플들을 벡터로 표현
$$\tilde{x}[n] = \tilde{x}[n + kN]$$

$$ \mathbf{x} = \begin{bmatrix} \tilde{x}[0] \\ \tilde{x}[1] \\ \vdots \\ \tilde{x}[N-1] \end{bmatrix}$$

- $N$ 차원 복소 공간의 한 벡터를 표현할 때 사용하는 정규기저 벡터를 변경하는 것




#### 2차원 실수 공간에서 벡터의 표현

- 2차원 실수 공간 $\mathbb{R}^2$에서 $\mathbf{u}_0 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$,   $\mathbf{u}_1 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}$이라고 하자.

$$\begin{bmatrix} 2 \\ 3 \end{bmatrix} = \begin{bmatrix} 2 \\ 0 \end{bmatrix} + \begin{bmatrix} 0 \\ 3 \end{bmatrix} = 2\begin{bmatrix} 1 \\ 0 \end{bmatrix} + 3 \begin{bmatrix} 0 \\ 1 \end{bmatrix} = 2 \mathbf{u}_0 + 3 \mathbf{u}_1$$

$$\begin{bmatrix} x_0 \\ x_1 \end{bmatrix} = \begin{bmatrix} x_0 \\ 0 \end{bmatrix} + \begin{bmatrix} 0 \\ x_1 \end{bmatrix} = x_0\begin{bmatrix} 1 \\ 0 \end{bmatrix} + x_1 \begin{bmatrix} 0 \\ 1 \end{bmatrix} = x_0 \mathbf{u}_0 + x_1 \mathbf{u}_1$$

- $\mathbf{u}_0$와 $\mathbf{u}_1$은 $\mathbb{R}^2$의 직교 정규 기저 벡터 (orthonormal basis vectors)
  - $\mathbf{u}_0$와 $\mathbf{u}_1$은 서로 직교
  - $|\mathbf{u}_0| = \sqrt{1^2 + 0^2} = 1$
  - $|\mathbf{u}_1 | = \sqrt{0^2 + 1^2} = 1$

<div class="alert alert-block alert-success">

#### 2차원 벡터의 내적

2차원 실수 공간 $\mathbb{R}^2$의 두 벡터 $\mathbf{x} = \begin{bmatrix} x_0 \\ x_1 \end{bmatrix}$과 $\mathbf{y} = \begin{bmatrix} y_0 \\ y_1 \end{bmatrix}$의 내적은 다음과 같이 정의한다.

$$\left\langle \mathbf{x}, \mathbf{y} \right\rangle = \left\langle \begin{bmatrix} x_0 \\ x_1 \end{bmatrix}, \begin{bmatrix} y_0 \\ y_1 \end{bmatrix} \right\rangle = x_0y_0 + x_1 y_1$$
</div>

#### 벡터의 내적의 기하학적 의미

$$\left\langle \mathbf{x}, \mathbf{y} \right\rangle = |\mathbf{x}| \; |\mathbf{y}| \cos\theta,$$
여기에서 $\theta$는 두 벡터가 이루는 각

- 두 벡터가 직교하면, $\theta = \frac{\pi}{2}$ 또는 $\theta = \frac{3\pi}{2}$이고  $\cos \frac{\pi}{2} = 0$이고 $\cos  \frac{3\pi}{2} =0$이므로, 직교하는 두 벡터의 내적은 0

#### 벡터의 크기

$$| \mathbf{x} | = \sqrt{x_0^2 + x_1^2} = \sqrt{\left\langle \mathbf{x}, \mathbf{x} \right\rangle} $$


#### 2차원 실수 공간의 직교 정규 기저 벡터 $\mathbf{u}_0, \mathbf{u}_1$

-  $ \left\langle \mathbf{u}_0, \mathbf{u}_0 \right\rangle = \left\langle \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \begin{bmatrix} 1 \\ 0 \end{bmatrix} \right\rangle = 1\cdot 1 + 0 \cdot 0 = 1$ 

<br>

-  $ \left\langle \mathbf{u}_0, \mathbf{u}_1 \right\rangle = \left\langle \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \begin{bmatrix} 0 \\ 1 \end{bmatrix} \right\rangle = 1\cdot 0 + 0 \cdot 1 = 0$

<br>

-  $ \left\langle \mathbf{u}_1, \mathbf{u}_0 \right\rangle = \left\langle \begin{bmatrix} 0 \\ 1 \end{bmatrix}, \begin{bmatrix} 1 \\ 0 \end{bmatrix} \right\rangle = 0\cdot 1 + 1 \cdot 0 = 0$

<br>

-  $ \left\langle \mathbf{u}_1, \mathbf{u}_1 \right\rangle = \left\langle \begin{bmatrix} 0 \\ 1 \end{bmatrix}, \begin{bmatrix} 0 \\ 1 \end{bmatrix} \right\rangle = 0\cdot 0 + 1 \cdot 1 = 1$



- $\mathbf{u}_0$와 $\mathbf{u}_1$은 $\mathbb{R}^2$의 직교 정규 기저 벡터 (orthonormal basis vectors)
  - $\mathbf{u}_0$와 $\mathbf{u}_1$은 서로 직교
  - $|\mathbf{u}_0| = \sqrt{1} = 1$
  - $|\mathbf{u}_1 | = \sqrt{1} = 1$

#### 직교 정규 기저 벡터를 이용한 벡터의 표현


$$\begin{bmatrix} 2 \\ 3 \end{bmatrix} = \begin{bmatrix} 2 \\ 0 \end{bmatrix} + \begin{bmatrix} 0 \\ 3 \end{bmatrix} = 2\begin{bmatrix} 1 \\ 0 \end{bmatrix} + 3 \begin{bmatrix} 0 \\ 1 \end{bmatrix} = 2 \mathbf{u}_0 + 3 \mathbf{u}_1$$

$$\mathbf{x} = \begin{bmatrix} x_0 \\ x_1 \end{bmatrix} = \begin{bmatrix} x_0 \\ 0 \end{bmatrix} + \begin{bmatrix} 0 \\ x_1 \end{bmatrix} = x_0\begin{bmatrix} 1 \\ 0 \end{bmatrix} + x_1 \begin{bmatrix} 0 \\ 1 \end{bmatrix} = x_0 \mathbf{u}_0 + x_1 \mathbf{u}_1$$

그런데

$<\mathbf{u}_0,  \mathbf{x}> = \left\langle \begin{bmatrix} 1 \\ 0 \end{bmatrix}, \begin{bmatrix} x_0 \\ x_1 \end{bmatrix} \right\rangle = x_0$, 
$<\mathbf{u}_1,  \mathbf{x}> = \left\langle \begin{bmatrix} 0 \\ 1 \end{bmatrix}, \begin{bmatrix} x_0 \\ x_1 \end{bmatrix} \right\rangle = x_1$이므로,

$$\mathbf{x} = \begin{bmatrix} x_0 \\ x_1 \end{bmatrix} = x_0 \mathbf{u}_0 + x_1 \mathbf{u}_1 = <\mathbf{u}_0,  \mathbf{x}> \mathbf{u}_0 + <\mathbf{u}_1,  \mathbf{x}> \mathbf{u}_1$$


- 또 다른 직교정규 기저 벡터: $\mathbf{w}_0, \mathbf{w}_1$

$$\mathbf{w}_0 = \begin{bmatrix} \frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2}\end{bmatrix}, \;\; \mathbf{w}_1 = \begin{bmatrix} -\frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2}\end{bmatrix} $$

- 직교
  - $<\mathbf{w}_0, \mathbf{w}_1> = \left\langle \begin{bmatrix} \frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix},\begin{bmatrix} -\frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix} \right\rangle = - \frac{2}{4} + \frac{2}{4} =0, $
  - $<\mathbf{w}_1, \mathbf{w}_0> = \left\langle \begin{bmatrix} -\frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix},\begin{bmatrix} \frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix} \right\rangle = - \frac{2}{4} + \frac{2}{4} =0, $
- 단위 벡터
  - $|\mathbf{w}_0| = \sqrt{<\mathbf{w}_0, \mathbf{w}_0>} = \sqrt{\left\langle \begin{bmatrix} \frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix},\begin{bmatrix} \frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix} \right\rangle} = \sqrt{ \frac{2}{4} + \frac{2}{4}} =1, $
  - $|\mathbf{w}_1| = \sqrt{<\mathbf{w}_1, \mathbf{w}_1>} = \sqrt{\left\langle \begin{bmatrix} -\frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix},\begin{bmatrix} -\frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix} \right\rangle} = \sqrt{ \frac{2}{4} + \frac{2}{4}} =1, $


#### 직교 정규 기저 벡터 $\mathbf{w}_0, \mathbf{w}_1$을 이용한 벡터의 표현


$$\begin{bmatrix} 2 \\ 3 \end{bmatrix} = a \mathbf{w}_0 + b \mathbf{w}_1,$$

$$a = \left\langle\mathbf{w}_0, \begin{bmatrix} 2 \\ 3 \end{bmatrix}\right\rangle = \left\langle \begin{bmatrix} \frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix}, \begin{bmatrix} 2 \\ 3 \end{bmatrix} \right\rangle =  \frac{5\sqrt{2}}{2},$$

</br>

$$b = \left\langle\mathbf{w}_1, \begin{bmatrix} 2 \\ 3 \end{bmatrix}\right\rangle = \left\langle \begin{bmatrix} -\frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix}, \begin{bmatrix} 2 \\ 3 \end{bmatrix} \right\rangle =  \frac{\sqrt{2}}{2},$$

따라서 
$$\begin{bmatrix} 2 \\ 3 \end{bmatrix} = \left\langle\mathbf{w}_0, \begin{bmatrix} 2 \\ 3 \end{bmatrix}\right\rangle \mathbf{w}_0 + \left\langle\mathbf{w}_1, \begin{bmatrix} 2 \\ 3 \end{bmatrix}\right\rangle \mathbf{w}_1 = \frac{5\sqrt{2}}{2} \mathbf{w}_0 + \frac{\sqrt{2}}{2} \mathbf{w}_1,$$

$$\mathbf{x} = \begin{bmatrix} x_0 \\ x_1 \end{bmatrix} 
= \left\langle\mathbf{u}_0, \mathbf{x} \right\rangle \mathbf{u}_0  + \left\langle\mathbf{u}_1, \mathbf{x} \right\rangle \mathbf{u}_1 
$$

</br>

$$\mathbf{x} = \begin{bmatrix} x_0 \\ x_1 \end{bmatrix} 
= \left\langle\mathbf{w}_0, \mathbf{x} \right\rangle \mathbf{w}_0  + \left\langle\mathbf{w}_1, \mathbf{x} \right\rangle \mathbf{w}_1 
$$

</br>

$$\left\langle\mathbf{w}_0, \mathbf{x} \right\rangle = \left\langle  \begin{bmatrix} \frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix}, \begin{bmatrix} x_0 \\ x_1 \end{bmatrix}\right\rangle = \frac{(x_0+x_1)\sqrt{2}}{2}$$

</br>

$$\left\langle\mathbf{w}_1, \mathbf{x} \right\rangle = \left\langle  \begin{bmatrix} -\frac{\sqrt{2}}{2} \\  \frac{\sqrt{2}}{2} \end{bmatrix}, \begin{bmatrix} x_0 \\ x_1 \end{bmatrix}\right\rangle = \frac{(-x_0+x_1)\sqrt{2}}{2}$$

### 직교 정규 기저 벡터를 이용한 $N$ 차원 실수 벡터의 표현

$\mathbf{u}_0$, $\mathbf{u}_1, \cdots, \mathbf{u}_{N-1}$을 $\mathbb{R}^N$ 공간의 직교 정규 기저 벡터(orthonormal basis vectors)라고 하면, 다음 식이 성립한다.

\begin{align*} \langle\mathbf{u}_i, \mathbf{u}_i\rangle = 1, &\mbox{ for } i=0,1,\cdots,N-1, \\
    \langle\mathbf{u}_i, \mathbf{u}_j\rangle = 0, &\mbox{ for } i \ne j, i, j=0,1,\cdots,N-1, 
\end{align*}

여기에서 $\langle \cdot, \cdot \rangle$는 두 벡터의 내적(inner product)으로 다음과 같이 정의된다.



<div class="alert alert-block alert-success">
  
    
**$N$ 차원 실수 벡터 공간 $\mathbb{R}^N$의 두 벡터 $\mathbf{a}$와 $\mathbf{b}$의 내적**
    
$\mathbf{a}=\begin{bmatrix}a_0& a_1& \cdots& a_{N-1}\end{bmatrix}^\intercal$와 $\mathbf{b}=\begin{bmatrix}b_0& b_1& \cdots& b_{N-1}\end{bmatrix}^\intercal$를 $N$ 차원 실수 벡터 공간 $\mathbb{R}^N$의 두 벡터라 하자. 여기에서 $\intercal$는 행렬의 전치(transpose)를 의미한다. $\mathbf{a}$와 $\mathbf{b}$의 내적은 다음과 같이 정의된다.
        \begin{equation}
        \langle \mathbf{a}, \mathbf{b} \rangle = \mathbf{a}^\intercal\mathbf{b}= {\displaystyle \sum_{i=0}^{N-1} a_i b_i=a_0 b_0 + a_1b_1
            + \cdots + a_{N-1} b_{N-1}}.
        \end{equation}

</div>

\begin{align}
    X_0 &= \langle\mathbf{v}_0, \mathbf{x} \rangle = 3.5, \\
    X_1 &= \langle\mathbf{v}_1, \mathbf{x} \rangle = 0.5, \\

그러면 $\mathbf{x}$를 다음과 같이 쓸 수 있다. 
$$
\mathbf{x} = \begin{bmatrix}1 \\ 3 \\ 4 \\ -1\end{bmatrix} = 3.5\mathbf{v}_0 + 0.5\mathbf{v}_1 + (-3.5)\mathbf{v}_2 + 1.5\mathbf{v}_3.
$$

```python
sv0.dot(x)*sv0 + sv1.dot(x)*sv1 + sv2.dot(x)*sv2 + sv3.dot(x)*sv3
```

$X_k = \langle\mathbf{v}_k, \mathbf{x} \rangle, \; k=0, 1, 2,3,$이라고 하자. $X_k$는 다음과 같이 구할 수 있다.

\begin{align}
    X_0 &= \langle\mathbf{v}_0, \mathbf{x} \rangle = 3.5, \\
    X_1 &= \langle\mathbf{v}_1, \mathbf{x} \rangle = 0.5, \\
    X_2 &= \langle\mathbf{v}_2, \mathbf{x} \rangle = -3.5, \\
    X_3 &= \langle\mathbf{v}_3, \mathbf{x} \rangle = 1.5.
\end{align}

벡터 $\mathbf{X}$를 $\mathbf{X} = \begin{bmatrix}X_0 \\ X_1 \\ X_2 \\ X_3\end{bmatrix}$으로 정의하면, $X_k$를 개별적으로 구하는 대신에 다음과 같이 행렬과 벡터의 곱으로 표현할 수 있다.

$$
\mathbf{X} = \begin{bmatrix}X_0 \\ X_1 \\ X_2 \\ X_3\end{bmatrix} =

### 직교 정규 기저 벡터를 이용한 $N$ 차원 복소 벡터의 표현

- 2차원 복소벡터의 예
$$
\begin{bmatrix} 1+ 2j \\ -2 - 3j \end{bmatrix} 
= \begin{bmatrix} 1+ 2j \\ 0 \end{bmatrix} + \begin{bmatrix} 0 \\ -2 - 3j \end{bmatrix}
= (1+ 2j)\begin{bmatrix} 1 \\ 0 \end{bmatrix} + (-2 - 3j)\begin{bmatrix} 0 \\ 1 \end{bmatrix}
= (1+ 2j)\mathbf{u}_0  + (-2 - 3j)\mathbf{u}_1 $$

$\mathbb{R}^N$ 공간과 마찬가지로 $\mathbb{C}^N$ 공간에도 $\mathbf{u}_0$, $\cdots$, $\mathbf{u}_{N-1}$ 외에도 수많은 직교 정규 기저 벡터들이 존재한다. 

$N$ 개의 벡터 $\mathbf{w}_0$, $\cdots$, $\mathbf{w}_{N-1}$도 직교 정규 기저 벡터라고 하면, $\mathbf{x}$를 다음과 같이 표현할 수도 있다.
\begin{equation} \begin{array}{ll}
\mathbf{x} &= \langle\mathbf{u}_0, \mathbf{x} \rangle\mathbf{u}_0 +
\cdots + \langle\mathbf{u}_{N-1}, \mathbf{x}\rangle\mathbf{u}_{N-1} \\
&= \langle\mathbf{w}_0, \mathbf{x}\rangle\mathbf{w}_0 + \cdots +
\langle\mathbf{w}_{N-1}, \mathbf{x}\rangle\mathbf{w}_{N-1}.
\end{array}
\label{x-in-v1-vN}
\end{equation}

$\mathbb{C}^N$ 공간의 두 벡터 $\mathbf{a}$와 $\mathbf{b}$의 내적은 다음과 같이 정의된다.
\begin{equation}
\langle \mathbf{a}, \mathbf{b} \rangle = {\mathbf{a}^\ast}^\intercal\mathbf{b}= {\displaystyle \sum_{i=0}^{N-1} a_i^\ast b_i=a_0^\ast b_0 + a_1^\ast b_1
    + \cdots + a_{N-1}^\ast b_{N-1}},
\end{equation}
여기에서 $^\ast$는 복소수의 공액(complex conjugate)을 의미한다.

### 이산 푸리에 급수


$\tilde{x}[n]$을 주기가 $N$인 이산 복소 주기 신호라고 하자. 

$$
\tilde{x}[n] = \tilde{x}[n + rN], \;\; r \mbox{는 정수.}
$$

$\tilde{x}[n]$의 한 주기 동안의 샘플들을 모아서 다음과 같이 벡터로 만든 것을 $\mathbf{x}$라고 하자.

$$
\mathbf{x} = \begin{bmatrix}\tilde{x}[0] \\ \tilde{x}[1] \\ \vdots \\ \tilde{x}[N-1]\end{bmatrix}.
$$

이 벡터를 다음과 같이 직교 정규 벡터들의 선형 결합으로 표현할 수 있다.
\begin{equation}
\mathbf{x} = \tilde{x}[0] \mathbf{u}_0 + \cdots + \tilde{x}[N-1] \mathbf{u}_{N-1},
\end{equation}
여기에서 
\begin{equation}
\mathbf{u}_k = \begin{bmatrix}0 \\ \vdots \\ 0 \\ 1 \\  0 \\ \vdots \\ 0\end{bmatrix}
\end{equation}
이다. 즉, $\mathbf{u}_k$는 k 번째 원소만 1이고 나머지는 0인 벡터이다.
이렇게 표현되는 $\mathbf{x}$의 $k$ 번째 원소인 $\tilde{x}[k]$는 시간 영역에서 $k$ 번째 샘플로 이해할 수 있다.


그리고 $\mathbf{x}$를 다음과 같이 또 다른 직교 정규 기저 벡터들의 선형 결합으로도 표현할 수 있다.

$$
\mathbf{x} = \langle \mathbf{w}_0, \mathbf{x} \rangle \mathbf{w}_0 + \cdots + \langle \mathbf{w}_{N-1}, \mathbf{x} \rangle \mathbf{w}_{N-1},
$$

여기에서  $W_N = e^{-j\frac{2\pi}{N}}$이고, $\mathbf{w}_k$는 다음과 같다.
\begin{equation} 
\mathbf{w}_k =  {\displaystyle \frac{1}{\sqrt{N}}\begin{bmatrix}{W_N^\ast}^0& {W_N^\ast}^{k}& {W_N^\ast}^{2k}& \cdots& {W_N^\ast}^{(N-1)k}\end{bmatrix}^\intercal}, k=0,1,\cdots,N-1.
\end{equation}

이제 $X_k=\langle \mathbf{w}_k, \mathbf{x} \rangle$라고 하면, $X_k$는 다음과 같이 쓸 수 있다.
\begin{align*}
    X_k &= \langle \mathbf{w}_k, \mathbf{x} \rangle \\[2mm]
    &= {\mathbf{w}_k^\ast}^\intercal \mathbf{x}  \\[2mm]
    &=  {\displaystyle \frac{1}{\sqrt{N}}\begin{bmatrix}W_N^0& W_N^k& W_N^{2k}& \cdots& W_N^{(N-1)k}\end{bmatrix} \begin{bmatrix}\tilde{x}[0]& \tilde{x}[1]& \cdots& \tilde{x}[N-1]\end{bmatrix}^\intercal } \\[4mm]
    &= {\displaystyle \frac{1}{\sqrt{N}}\left(  W_N^0\tilde{x}[0] +W_N^{k}\tilde{x}[1] + \cdots +   W_N^{(N-1)k} \tilde{x}[N-1]\right) }\ \\[4mm]
    %&= \hush{{\displaystyle \frac{1}{\sqrt{N}}\left[ W_N^0\tilde{x}[0] + W_N^{-k}\tilde{x}[1] + \cdots +   W_N^{-(N-1)k}\tilde{x}[N-1]\right] }}\\[4mm] 
    &= {{\displaystyle \frac{1}{\sqrt{N}}\sum_{n=0}^{N-1} W_N^{nk} \tilde{x}[n] }}\\[2mm]
    &= {\displaystyle \frac{1}{\sqrt{N}}\sum_{n=0}^{N-1} e^{-j\frac{2\pi}{N}nk} \tilde{x}[n]. } 
\end{align*}
